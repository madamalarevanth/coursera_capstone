Neighborhoods in Toronto Data Set

import packages

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Scrape Wikipidia Page

```python
page = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
table = soup.find('tbody')
rows = table.select('tr')
row = [r.get_text() for r in rows]
```

Clean the DataFrame

```python
df = pd.DataFrame(row)
df1 = df[0].str.split('\n', expand=True)
df2 = df1.rename(columns=df1.iloc[0])
df3 = df2.drop(df2.index[0])
df3.head()
```

Remove the Not Assigned Rows 

```python
df=df3[df3['Borough']!='Not assigned']
df
```

Join the different Neighbour hood under the same borrrow

```python
df5 = df.groupby(['Postcode', 'Borough'], sort = False).agg(','.join)
df5.reset_index(inplace = True)
df5
```

in this case there is only one neighbour hood wich isnt assigned with a borrow name so directly add the neighbourhood as borough

```python
df6  =  df5.replace("Not assigned", "Queen's Park")
df6.head()
```
